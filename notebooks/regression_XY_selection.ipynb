```python
import pandas as pd
import os 
from pathlib import Path

project_dir = Path(os.path.abspath('')).resolve().parent
data_dir = project_dir.joinpath("data")

pd.set_option('display.max_rows', None)
```

# Load Data

```python
survey_file_path = data_dir.joinpath("output/clean_anon_survey.csv")
questions_file_path = data_dir.joinpath("questionlist.csv")
survey_df = pd.read_csv(survey_file_path, sep=';')
questions_df = pd.read_csv(questions_file_path, sep=';')
```

# Check completeness of responses

```python
relevant_fields = questions_df.loc[questions_df['Group'] != 'Metadata', 'Code']

response_counts = survey_df.count().sort_values(ascending=False)
response_counts = response_counts[response_counts.index.isin(relevant_fields)]
response_counts_df = response_counts[response_counts > 70].to_frame(name='response_counts')

questions_text = questions_df[['Code', 'ShortQuestion']].set_index('Code')
response_counts_df = pd.merge(response_counts_df, questions_text,how='left', left_index=True, right_index=True)
response_counts_df['missing percentage'] =  round(100 -(response_counts_df['response_counts'] / survey_df.shape[0]) * 100)
response_counts_df

```
